```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction import DictVectorizer
```

```python
df= pd.read_csv("names.csv")
df=df.dropna()
df
```

```python
df.columns
```

```python
df.gender.replace({"Male":1,"Female":0},inplace=True)
df
```

```python
df.gender.unique()
```

```python
df=df[df['gender']!='Other']
```

```python
df.gender.unique()
```

```python
names_df=df['name']
```

```python
names_df
```

```python
cv=CountVectorizer()
X=cv.fit_transform(names_df)
```

```python
X
```

```python
cv.get_feature_names()
```

```python
from sklearn.model_selection import train_test_split
```

```python
y=df.gender
```

```python
y=y.astype(int)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y, test_size=.2, random_state=12)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
clf= MultinomialNB()
```

```python
clf.fit(X_train,y_train)
```

```python
clf.score(X_test,y_test)
```

```python
type(X_train[0])
```

```python
clf.fit(X_train,y_train)
clf.score(X_test,y_test)
```

```python
test_df=pd.read_csv("test.csv")
```

```python
test_df['name']
```

```python
test_df=test_df['name'].dropna()
```

```python
vect=cv.transform(test_df).toarray()
```

```python
vect
```

```python
result=clf.predict(vect)
```

```python
result_df=pd.concat([test_df, pd.Series(result)], axis=1).reset_index()
```
